```python
from typing import Dict, Set, Optional

from pyprojroot import here
```

```python
def addRock(coordsList, cave: Dict[int, Set[int]]):
    currentX, currentY = coordsList[0]

    for nextX, nextY in coordsList[1:]:
        moveX = nextX - currentX
        moveY = nextY - currentY
        
        # moving along x-axis
        if moveX != 0:
            row = currentY
            start, stop = sorted([currentX, nextX])
            cols = range(start, stop + 1)

            for col in cols:
                cave.setdefault(col, set()).add(row)

        # moving along y-axis
        if moveY != 0:
            col = currentX
            start, stop = sorted([currentY, nextY])
            rows = range(start, stop + 1)
            cave.setdefault(col, set()).update(list(rows))

        currentX = nextX
        currentY = nextY


def buildCave(lines):
    rockFormations = {}
    for line in lines:
        addRock(line, rockFormations)

    return rockFormations


def printCave(cave, sand):
    cols = sorted(cave.keys())
    allRows = [row for rows in cave.values() for row in rows] + [0]
    rows = list(range(min(allRows), max(allRows) + 1))

    for row in rows:
        line = ''

        for col in cols:

            # starting point
            if (col, row) == (500, 0):
                line += '+'

            # sand
            elif (col, row) in sand:
                line += 'o'

            # rock
            elif cave.get(col) and row in cave.get(col):
                line += "#"

            # air
            else:
                line += "\u00B7"

        print(line)
```

```python
def dropSand(cave: dict, sand: set, col=500, row=0):

    # into the abyss
    if row > max(cave[col]):
        return False

    # if space below
    if not row + 1 in cave[col]:
        return dropSand(cave, sand, col, row + 1)

    # into the abyss left
    if not cave.get(col - 1):
        return False

    # if space diagonal left
    if not row + 1 in cave[col - 1]:
        return dropSand(cave, sand, col - 1, row + 1)

    # into the abyss right
    if not cave.get(col + 1):
        return False

    # if space diagonal right
    if not row + 1 in cave[col + 1]:
        return dropSand(cave, sand, col + 1, row + 1)

    # stop
    cave[col].add(row)
    sand.add((col, row))

    return True


def flowSand(cave):
    sand = set()
    sandCount = 0
    flowingIntoTheAbyss = False

    while not flowingIntoTheAbyss:
        flowingIntoTheAbyss = not dropSand(cave, sand)

        if not flowingIntoTheAbyss:
            sandCount += 1
    
    printCave(cave, sand)
    return sandCount
```

```python
path = here('./14/input.txt')

with open(path, 'r') as fp:
    rawLines = [line.strip() for line in fp.readlines()]
    lines = [[[int(char) for char in coords.split(',')] for coords in line.split(' -> ')] for line in rawLines]
    
cave = buildCave(lines)
flowSand(cave)
```
